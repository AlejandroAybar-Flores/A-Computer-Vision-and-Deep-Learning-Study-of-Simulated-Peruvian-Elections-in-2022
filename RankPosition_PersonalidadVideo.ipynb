```python
import pandas as pd
import itertools
import numpy as np
import statistics

#Funciones

def ordenar(df):
    lis_ocean=["O","C","E","A","N"]
    lis_nombre = list(df.columns)
    lis_df=[]
    for i in lis_nombre:
        lis_column = df[i].tolist()
        dic_per1 = itertools.zip_longest(lis_ocean, lis_column, fillvalue=None)
        dic_per = dict(dic_per1)
        sortedDict = sorted(dic_per.items(), key=lambda x: x[1], reverse=True)
        lis_df.append(sortedDict)
        
    df1 = pd.DataFrame(lis_df)
    df1=df1.transpose()
    df1.columns = list(df.columns)
    return df1

def df_ord(fila):
    rasgo=[]
    for f in fila:
        tupla=f[0]
        rasgo.append(tupla)
    dfr = pd.Series(rasgo).value_counts().sort_index().reset_index().reset_index(drop=True)
    dfr.columns = ['elemento', 'frecuencia']
    dfr = dfr.sort_values('frecuencia',ascending=False)
    l_dfr = list(dfr['elemento'])
    return  l_dfr
    
def pervid(df):
    per_vid = df.to_numpy().tolist()
    per_vid=list(per_vid)
    
    ocean_completo = []

    # Fila 0
    l_ord0 = df_ord(per_vid[0])
    mayor0 = l_ord0[0]
    ocean_completo.append(mayor0)

    # Fila 1
    l_ord1 = df_ord(per_vid[1])
    mayor1 = l_ord1[0]
    ocean_completo.append(mayor1)

    # Fila 2
    l_ord2 = df_ord(per_vid[2])
    mayor2 = l_ord2[0]
    ocean_completo.append(mayor2)

    # Fila 3
    l_ord3 = df_ord(per_vid[3])
    mayor3 = l_ord3[0]
    ocean_completo.append(mayor3)

    # Fila 4
    l_ord4 = df_ord(per_vid[4])
    mayor4 = l_ord4[0]
    ocean_completo.append(mayor4)
    
    return ocean_completo

def prom_fila(rasgo,fila):
    ele_suma=[]
    for i in fila:
        if i[0]==rasgo:
            ele_suma.append(i[1])
    promedio = statistics.mean(ele_suma)
    return promedio

def promedio(per_vid,df_per):
    ras_todos = df_per.to_numpy().tolist()
    ras_todos=list(ras_todos)
    video_per=[]
    
    # Fila 0
    rasgo0=per_vid[0]
    fila0=ras_todos[0]
    prome0 = prom_fila(rasgo0,fila0)
    video_per.append(prome0)
    
    # Fila 1
    rasgo1=per_vid[1]
    fila1=ras_todos[1]
    prome1 = prom_fila(rasgo1,fila1)
    video_per.append(prome1)
    
    # Fila 2
    rasgo2=per_vid[2]
    fila2=ras_todos[2]
    prome2 = prom_fila(rasgo2,fila2)
    video_per.append(prome2)
    
    # Fila 3
    rasgo3=per_vid[3]
    fila3=ras_todos[3]
    prome3 = prom_fila(rasgo3,fila3)
    video_per.append(prome3)
    
    # Fila 4
    rasgo4=per_vid[4]
    fila4=ras_todos[4]
    prome4 = prom_fila(rasgo4,fila4)
    video_per.append(prome4)
    
    return per_vid,video_per

# Función que une todas las funciones

def calcular_pervid(nro):
    def df_video(video):
        groups = df.groupby(df.Video)
        df_vid = groups.get_group(video)
        df_vid = df_vid.drop(['Candidato'], axis=1)
        df_vid = df_vid.drop(['Video'], axis=1)
        lis_img1 = list(df_vid["img"])
        lis_img=[]
        for i in lis_img1:
            texto=i.split("_")
            parte1=texto[0]
            parte2=texto[2]
            texto=parte1+"_"+parte2
            lis_img.append(texto)
        df_vid = df_vid.drop(['img'], axis=1)
        df_vid = df_vid.transpose()
        df_vid.columns = lis_img
        return df_vid
    
    lista_vide=lista_video[nro]
    df_video = df_video(lista_vide)
    df_per = ordenar(df_video)
    per_vid = pervid(df_per)
    promedio_rasgos = promedio(per_vid,df_per)
    
    lista_resultado=[]
    lista_resultado.append(lista_vide)
    lista_resultado.append(promedio_rasgos)
    return lista_resultado
```
