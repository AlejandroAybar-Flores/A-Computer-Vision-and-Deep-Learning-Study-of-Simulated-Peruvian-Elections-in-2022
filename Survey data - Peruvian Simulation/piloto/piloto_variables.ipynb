```python
from collections import Counter

def listacc():
    listac=[]
    listaf=[] 
    listacat=[]
    for ind in range(0,13):
        datos=list(df_cat[var_cat[ind]])
        cat = list(Counter(datos).keys())
        fre = list(Counter(datos).values())
        cant=len(fre)
        lista_cat=[]
        for j in range(0,cant):
            lista_cat.append(var_cat[ind])
        listacat.extend(lista_cat)
        
        listac.extend(cat)
        listaf.extend(fre)
    return listac

def listaff():
    listac=[]
    listaf=[] 
    listacat=[]
    for ind in range(0,13):
        datos=list(df_cat[var_cat[ind]])
        cat = list(Counter(datos).keys())
        fre = list(Counter(datos).values())
        cant=len(fre)
        lista_cat=[]
        for j in range(0,cant):
            lista_cat.append(var_cat[ind])
        listacat.extend(lista_cat)
        
        listac.extend(cat)
        listaf.extend(fre)
    return listaf

def listacatt():
    listac=[]
    listaf=[] 
    listacat=[]
    for ind in range(0,13):
        datos=list(df_cat[var_cat[ind]])
        cat = list(Counter(datos).keys())
        fre = list(Counter(datos).values())
        cant=len(fre)
        lista_cat=[]
        for j in range(0,cant):
            lista_cat.append(var_cat[ind])
        listacat.extend(lista_cat)
        
        listac.extend(cat)
        listaf.extend(fre)
    return listacat
```

```python
listacat=listacatt()
listaetiq=listacc()
listafreq=listaff()
df_catt = pd.DataFrame(list(zip(listacat, listaetiq,listafreq)), 
               columns =['categoria', 'etiqueta','frecuencia'])
df_catt.index=listacat
df_catt
```
